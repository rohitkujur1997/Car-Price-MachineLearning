```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
catog=['Make', 'Model', 'Trim', 'Type']
newdataset=pd.get_dummies(dataset[catog],drop_first=True)  
```

```python
newdataset.head(5)
```

```python
dataset.drop(catog,axis=1,inplace=True)
```

```python
dataset=pd.concat([dataset,newdataset],axis=1)
```

```python
dataset.head(5)
```

```python
from sklearn.preprocessing import scale
```

```python
dataset['Mileage']=scale(dataset['Mileage'])
```

```python
dataset
```

```python
dataset.info()
```

```python
x = dataset.drop('Price',axis=1)
y = dataset['Price']
```

**Splitting**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=69)
```

**Training** **Model**

```python
#from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
```

```python
#d_tree = DecisionTreeClassifier()
linear=LinearRegression()
```

```python
linear.fit(x_train,y_train)
```

**Predicting**

```python
linear_y_train_pred=linear.predict(x_train)
```

```python
linear_y_test_pred=linear.predict(x_test)
```

**Accuracy**

```python
from sklearn.metrics import accuracy_score,r2_score
```

```python
r2_score(y_true=y_train,y_pred=linear_y_train_pred)
```

```python
r2_score(y_true=y_test,y_pred=linear_y_test_pred)
```
